```python
# private and public attribute 
class People:
    def __init__(self, name, age):
        self.name = name
        self.__age = age # private 

    def get_age(self):
        return self.__age # private 

    def set_age(self, new_age):
        if new_age > 0:
            self.__age = new_age
            return True
        return False
```

```python
people1 = People("Ali", 45)

print(people1.get_age())  # 45

people1.set_age(46)
print(people1.get_age())  # 46

people1.set_age(30)       # if you want to update the value should be valide ortherwis print the old one
print(people1.get_age())  # Still 46
```

```python
#Creating a Parent Class
class Person:
    def __init__(self, name, age, email): # construction of a class with __init__
        self.name = name
        self.age = age
        self.email = email

    def display_info(self):
        return f"Name: {self.name}, Age: {self.age},  Email: {self.email}"

    def update_email(self, new_email):
        self.email = new_email
        return f"Email updated to {new_email}"
```

```python
# Creating a Child Class
class Student(Person):
    def __init__(self, name, age, email, student_id):
        super().__init__(name, age, email) # super let you access to parent class 
        self.student_id = student_id
        self.courses = []

    def enroll(self, course):
        self.courses.append(course)
        return f"{self.name} enrolled in {course}"
```

```python
student = Student("Alice Johnson", 20, "alice@school.edu", "S12345")

# Methods inherited from Person
print(student.display_info())
print(student.update_email("alice.j@school.edu"))
# Methods specific to Student
print(student.enroll("Mathematics"))
```

```python
# Multiple Child Classes
class Teacher(Person):
    def __init__(self, name, age, email, employee_id, subject):
        super().__init__(name, age, email)
        self.employee_id = employee_id
        self.subject = subject
        self.classes = []

    def assign_class(self, class_name):
        self.classes.append(class_name)
        return f"{self.name} assigned to teach {class_name}"

class Administrator(Person):
    def __init__(self, name, age, email, employee_id, department):
        super().__init__(name, age, email)
        self.employee_id = employee_id
        self.department = department

    def make_announcement(self, message):
        print(f"Announcement from {self.name}: {message}")        
```

```python
#Polymorphism and Flexible Code Design
#  Imagine you're building a drawing application 
# that needs to calculate the area of different shapes:
# This base class defines the interface — every shape should have an area() method
class Shape:
    def area(self):
        return 0

# now creating specific shape classes
class Rectangle(Shape):
    def __init__ (self , width, height):
        self.width = width
        self.height = height
    def area(self):
        return self.width * self.height

```

The Rectangle class overrides the area() method to provide rectangle-specific calculation logic. Each shape type will have its own formula for calculating area, but they all use the same method name. Let's add more shapes:

```python
# Create Circle
class Circle(Shape):
    def __init__ (self , radius):
        self.radius = radius
    def area(self):
        return  3.14159* self.radius**2

# Triangle shape 
class Triangle(Shape):
    def __init__ (self , base, height):
        self.base = base
        self.height = height 
    def area(self):
        return 0.5 * self.base * self.height
```

```python
#where polymorphism becomes powerful. You can write code that works with any shape without knowing the specific type:

def print_area(shape):
    result =shape.area()
    return f"Area: {result: .2f}"

rect = Rectangle(5, 10)
circle = Circle(7)
triangle = Triangle(6, 8)
print(f" Rectangl-{print_area(rect)}, Cricle-{print_area(circle)}, Triangle-{print_area(triangle)}")
```
